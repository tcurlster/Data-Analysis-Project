```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'col':np.random.randn(12000), 'target':np.random.randint(low = 0, high = 2, size=12000)})
df
```

```python
new_df = df.groupby('target').apply(lambda x: x.sample(n=5000)).reset_index(drop = True)
new_df
```

```python
new_df.target.value_counts()
```
